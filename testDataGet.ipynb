```python
import urllib
import requests
import string
import re
from bs4 import BeautifulSoup
```

```python
def getSoupByUrl(url):
    resUrl = urllib.parse.quote(url,safe=string.printable)
    page = urllib.request.urlopen(resUrl).read()
    soup = BeautifulSoup(page,'html.parser')
    return soup # soup.prettify()

def getAllStations(i):
    all_stationNames = []
    url = "https://baike.baidu.com/item/北京地铁"+ str(i) +"号线"
    soup = getSoupByUrl(url)
    all_stations = soup.find_all('a',attrs={"target":"_blank"},string=re.compile("站"))
    for s in all_stations:
        text = s.get_text()
        if len(text) < 10:
            all_stationNames.append(text)
    return all_stationNames
```

```python
all_lines= locals()

for i in range(1,11):
    all_lines[i] = getAllStations(i)
```
